```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

### Traitement de donnée

```python
dataset=pd.read_csv('voitures.csv')
dataset
```

#### En codage des données

```python

dataset["carburant"].replace("Diesel",0,inplace=True)
dataset["carburant"].replace("Essence",1,inplace=True)
dataset["transmission"].replace("Manuelle",0,inplace=True)
dataset["transmission"].replace("Automatique",1,inplace=True)
dataset
```

```python
# fonction encode marque
def encode_marque(marque):
    return len(marque)
```

```python
dataset["marque"]=dataset["marque"].apply(encode_marque)
```

```python
dataset
```

### Normalisation


```python
features = dataset[["km","annee","carburant","transmission","marque"]]
target = dataset[["prix"]]
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
def z_score_normalization(data):
    # Calcul des moyennes et des déviations standard de chaque caractéristique
    mean = np.mean(data, axis=0)
    std = np.std(data, axis=0)

    # Normalisation des données en utilisant la méthode Z-score
    data_normalized = (data - mean) / std

    return data_normalized
```

```python
x_train=z_score_normalization(x_train)
#x_test=z_score_normalization(x_test)
y_train=z_score_normalization(y_train)
#y_test=z_score_normalization(y_test)
```

### Modéle

```python
x_train=np.hstack((x_train,np.ones((x_train.shape[0],1))))
x_train
```

```python
def model_prediction(X, theta):
    return np.dot(X, theta)
```

##### fonction cout


```python
def fonction_cout(X, y, theta):
    m = len(y)
    predictions = model_prediction(X, theta)
    cost = (1/(2*m)) * np.sum((predictions - y) ** 2)
    return cost
```

```python
def gradient(X, y, theta):
    m = len(y)
    predictions = model_prediction(X, theta)
    errors = predictions - y
    grad = (1/m) * np.dot(X.T, errors)
    return grad
```

```python
def gradient_descent(X, y, theta, learning_rate=0.01, iterations=3000):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        gradients = gradient(X, y, theta)
        theta -= learning_rate * gradients
        cost = fonction_cout(X, y, theta)
        cost_history.append(cost)

    return theta, cost_history
```

```python
np.random.seed(42)
theta_initial = np.random.randn(x_train.shape[1], 1)
theta_optimal, cost_history = gradient_descent(x_train, y_train, theta_initial)
theta_optimal
```

```python
plt.figure()
plt.plot(range(len(cost_history)), cost_history, c='blue')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Convergence of Gradient Descent')
plt.show()
```

```python
x_test=np.hstack((x_test,np.ones((x_test.shape[0],1))))
predictions = model_prediction(x_test, theta_optimal)
predictions

```

```python
def coef_determination(y, pred):
    u = ((y - pred)**2).sum()
    v = ((y - y.mean())**2).sum()
    return 1 - u/v
```
